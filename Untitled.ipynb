# SVM and kernel methods

The support vector machine (SVM) tries to find a linear decision boundary that maximises the margin to the closest training examples - with the aim of minimising the generalisation error.

```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import iris dataset
iris = datasets.load_iris()
data = np.array(iris['data'])  # data
target = np.array(iris['target'])  # classifications
data.shape, target.shape
```

```python
# take the first two columns of the data to train on, and look to identify the 0th class
X = data[:, 1:3]
y = np.where(target == 2, 1, -1)
```

## Use Perceptron

```python
from ml.linear_classifier import Perceptron
```

```python
perceptron = Perceptron(random_state=1)
cost_per_iteration = perceptron.fit(X, y, n_iterations=500, learning_rate=0.001)
plt.plot(cost_per_iteration)
```

```python
from ml.utils import plot_decision_regions
plt = plot_decision_regions(X, y, classifier=perceptron)
plt.show()
```

## Use SVM

```python
# import support vector classification model
from sklearn.svm import SVC
```

```python
# First use linear kernel
svm = SVC(kernel='linear', C=1.0, random_state=1)
svm.fit(X, y)
plot_decision_regions(X, y, classifier=svm)
plt.show()
```

```python
# First use linear kernel
svm_nl = SVC(kernel='rbf', C=1.0, random_state=1, gamma=0.5)
svm_nl.fit(X, y)
plot_decision_regions(X, y, classifier=svm_nl)
plt.show()
```

```python
# try on the second class
y = np.where(target == 1, 1, -1)

svm_nl = SVC(kernel='rbf', C=1.0, random_state=1, gamma=5)
svm_nl.fit(X, y)
plot_decision_regions(X, y, classifier=svm_nl)
plt.show()
```
